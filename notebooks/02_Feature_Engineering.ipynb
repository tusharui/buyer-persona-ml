## 02 — Feature Engineering

Deriving customer-level RFM and behavioral features from raw transaction data.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))
from src.config import PROCESSED_FILES, RAW_DATA_PATH

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)
```

```python
df = pd.read_csv(RAW_DATA_PATH, parse_dates=["InvoiceDate"])
print(f"Shape: {df.shape}")
df.head(3)
```

### Build Customer Features

```python
from src.features import build_customer_features

cust = build_customer_features(df)
print(f"Customers: {cust.shape[0]}, Features: {cust.shape[1]}")
cust.head()
```

### Save Features

```python
cust.to_csv(PROCESSED_FILES["features"], index=False)
print(f"Saved: {PROCESSED_FILES['features']}")
```

### Feature Distributions

```python
num_cols = cust.select_dtypes(include=[np.number]).columns[:8]
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for ax, col in zip(axes.flat, num_cols):
    ax.hist(cust[col].dropna(), bins=30, color="steelblue", edgecolor="black")
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
sns.heatmap(cust.select_dtypes(include=[np.number]).corr(), annot=True, cmap="RdBu_r", center=0, fmt=".2f", square=True, linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```
